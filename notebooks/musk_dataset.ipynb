```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
import keras
from keras.models import Sequential
from keras.layers import Dense , Dropout
from keras.optimizers import Adam
from sklearn.metrics import f1_score, precision_score, recall_score
```

```python
mk = pd.read_csv('/content/gdrive/My Drive/Crd_intern/musk_csv.csv')
```

**Preprocessing**

```python
#checking the Null value
mk.isnull().sum()
```

```python
#dropping the cat Variable
mk.drop(columns=['ID'	,'molecule_name','conformation_name'],inplace=True)
```

```python
mk.shape
```

```python
mk=np.array(mk)
```

```python
X = mk[:,1:-1]
Y = mk[:,-1]
```

```python
Y.shape
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X, Y, random_state=15,test_size=0.2)
```

```python
y_test.shape
```

**Model_CNN**

```python
model = Sequential()
model.add(Dense(24, input_dim=165, activation='relu')) 
model.add(Dense(20, activation='relu'))
model.add(Dense(16, activation='relu'))
model.add(Dense(8, activation='relu'))
model.add(Dropout(.25))  #Regularization
model.add(Dense(1, activation='sigmoid')) 
```

```python
model.compile(loss=keras.losses.binary_crossentropy, optimizer="adam", metrics=['accuracy'])
```

```python
m=model.fit(x_train, y_train, epochs = 25, batch_size=32, validation_data=(x_test, y_test))
```

```python
score=model.evaluate(x_test,y_test,verbose=0)
print(score)
```

**Measuring** **Performance**

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.plot(m.history["val_acc"],label="Validation Accuracy")
plt.plot(m.history["acc"],label="Training Accuracy")
plt.legend()
plt.xlabel("epochs")
plt.ylabel("Accuracy")
plt.show()

plt.plot(m.history["val_loss"],label="Validation loss")
plt.plot(m.history["loss"],label="Training loss")
plt.legend()
plt.xlabel("epochs")
plt.ylabel("loss")
plt.show()
```

```python
print("f1_score:",f1_score(y_test,model.predict_classes(x_test),))
print("recall:",recall_score(y_test,model.predict_classes(x_test),))
print("Validation Loss:",score[0])
print("Validation Accuracy:",score[1])
```
